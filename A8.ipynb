```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
titanic=sns.load_dataset('titanic')
titanic.head()
```

```python
#plt.figure(figsize=(10,6))
sns.histplot(data=titanic,x='fare')
plt.title('distribution of fare')
plt.xlabel('fare')
plt.ylabel('number of people')
plt.grid(True)
plt.show()
```
