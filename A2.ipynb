```python

data = {
    'Student_ID': range(1, 11),
    'Math_Score': [88, 92, np.nan, 70, 65, 120, 75, 85, 78, np.nan],  # Note: missing and an unrealistic score (120)
    'Science_Score': [90, 85, 80, 70, 75, 95, 77, np.nan, 68, 85],     # Missing value
    'English_Score': [85, 80, 78, 72, 68, 65, 90, 92, 60, 70],
    'Attendance_Rate': [0.95, 0.9, 0.85, 0.8, 0.75, 0.7, np.nan, 0.85, 0.6, 0.65]  # Missing value
}

df = pd.DataFrame(data)
print(df)
```

```python
data={
    'id':range(1,11),
    'DSBDA':[88,87,24,np.nan,90,45,29,np.nan,67,89],
    'WT':[90,np.nan,78,67,44,78,34,np.nan,89,34],
    'AI':[np.nan,78,98,65,43,23,98,np.nan,89,67]
}

df=pd.DataFrame(data)
df.head(20)
```

```python
df['DSBDA'].fillna(df['DSBDA'].mean(),inplace=True)
df['WT'].fillna(df['WT'].mean(),inplace=True)
df['AI'].fillna(df['AI'].mean(),inplace=True)
```

```python
df.head(11)
```

```python
df=df[np.abs(zscore(df['WT']))<3]
```

```python
df['log1']=np.log(df['DSBDA'])
df['log2']=np.log(df['WT'])
df['log3']=np.log(df['AI'])
```
